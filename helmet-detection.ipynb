```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Load trained YOLOv8 model
model = YOLO("/content/drive/MyDrive/Deep Learning Practicals/helmetdata/best.pt")

# Class names
class_names = [
    'driver_with_helmet',
    'bike',
    'driver',
    'passenger_with_helemt',
    'passenger',
    'driver_without_helmet',
    'passenger_without_helemt'
]

# Class colors and names
class_colors = {
    0: ((0, 255, 0), "Green"),
    1: ((255, 0, 0), "Blue"),
    2: ((0, 0, 255), "Red"),
    3: ((255, 255, 0), "Cyan"),
    4: ((255, 0, 255), "Magenta"),
    5: ((0, 255, 255), "Yellow"),
    6: ((128, 0, 128), "Purple")
}

# Load image
image_path = "/content/drive/MyDrive/Deep Learning Practicals/helmetdata/data/train/images/BikesHelmets169_png.rf.1502252b18b399e727992aa8688fa172.jpg"
original_img = cv2.imread(image_path)
img_copy = original_img.copy()

# Predict
results = model.predict(source=image_path, imgsz=640, conf=0.25)
boxes = results[0].boxes

# Track seen classes for legend
seen_classes = {}

# Font
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 0.6
thickness = 2

# Draw boxes and add "Approved"/"Not Approved" where applicable
for box in boxes:
    cls_id = int(box.cls[0])
    conf = float(box.conf[0])
    class_name = class_names[cls_id]
    (color, color_name) = class_colors[cls_id]

    x1, y1, x2, y2 = map(int, box.xyxy[0])
    cv2.rectangle(img_copy, (x1, y1), (x2, y2), color, 2)

    # Add "Approved"/"Not Approved" if needed
    if class_name == "driver_with_helmet":
        cv2.putText(img_copy, "Approved", (x1, y2 + 20), font, font_scale, (0, 255, 0), thickness)
    elif class_name == "driver_without_helmet":
        cv2.putText(img_copy, "Not Approved", (x1, y2 + 20), font, font_scale, (0, 0, 255), thickness)

    # Add to legend (only once per class)
    if cls_id not in seen_classes:
        seen_classes[cls_id] = f"{class_name} ({conf*100:.1f}%) - {color_name}"

# Convert for display
original_rgb = cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB)
detected_rgb = cv2.cvtColor(img_copy, cv2.COLOR_BGR2RGB)

# Plot original and detected
plt.figure(figsize=(16, 8))

plt.subplot(1, 2, 1)
plt.imshow(original_rgb)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(detected_rgb)
plt.title("Detected Image with Status")
plt.axis("off")

# Add legend
legend_lines = [f"{i+1}. {entry}" for i, entry in enumerate(seen_classes.values())]
plt.figtext(0.5, 0.01, "\n".join(legend_lines),
            wrap=True, horizontalalignment='center', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from ultralytics import YOLO

# Load your trained YOLOv8 model
model = YOLO("/content/drive/MyDrive/Deep Learning Practicals/helmetdata/best.pt")

# Class names and color definitions
class_names = [
    'driver_with_helmet',
    'bike',
    'driver',
    'passenger_with_helemt',
    'passenger',
    'driver_without_helmet',
    'passenger_without_helemt'
]

class_colors = {
    0: ((0, 255, 0), "Green"),
    1: ((255, 0, 0), "Blue"),
    2: ((0, 0, 255), "Red"),
    3: ((255, 255, 0), "Cyan"),
    4: ((255, 0, 255), "Magenta"),
    5: ((0, 255, 255), "Yellow"),
    6: ((128, 0, 128), "Purple")
}

# Load and copy image
image_path = "/content/drive/MyDrive/Deep Learning Practicals/helmetdata/data/train/images/BikesHelmets172_png.rf.9a8d520687831eaca169efb458b2105e.jpg"
original_img = cv2.imread(image_path)
img_copy = original_img.copy()

# Predict
results = model.predict(source=image_path, imgsz=640, conf=0.25)
boxes = results[0].boxes

# For final label display (once per class)
seen_classes = {}

# Font setup
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 0.7
thickness = 2

# Loop over detections
for box in boxes:
    cls_id = int(box.cls[0])
    conf = float(box.conf[0])
    class_name = class_names[cls_id]
    color, color_name = class_colors[cls_id]
    x1, y1, x2, y2 = map(int, box.xyxy[0])

    # Draw bounding box
    cv2.rectangle(img_copy, (x1, y1), (x2, y2), color, 2)

    # Add "Approved"/"Not Approved" text for driver classes
    if class_name == "driver_with_helmet":
        cv2.putText(img_copy, "APPROVED", (x1, y2 + 25), font, font_scale, (0, 255, 0), thickness)
    elif class_name == "driver_without_helmet":
        cv2.putText(img_copy, "NOT APPROVED", (x1, y2 + 25), font, font_scale, (0, 0, 255), thickness)

    # Save class info for bottom legend
    if cls_id not in seen_classes:
        seen_classes[cls_id] = f"{class_name.upper()} - {color_name.upper()} ({conf * 100:.1f}%)"

# Convert BGR to RGB
original_rgb = cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB)
detected_rgb = cv2.cvtColor(img_copy, cv2.COLOR_BGR2RGB)

# Plotting
plt.figure(figsize=(16, 8))

plt.subplot(1, 2, 1)
plt.imshow(original_rgb)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(detected_rgb)
plt.title("Detected Image with Helmet Status")
plt.axis("off")

# Add class summary at bottom in bold
legend_lines = [f"{i+1}. {entry}" for i, entry in enumerate(seen_classes.values())]
plt.figtext(0.5, 0.01, "\n".join(legend_lines), wrap=True,
            horizontalalignment='center', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# Count class occurrences
from collections import Counter
class_ids = [int(box.cls[0]) for box in results[0].boxes]
counts = Counter(class_ids)

# Plot
labels = [class_names[i] for i in counts.keys()]
sizes = list(counts.values())

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Detected Object Distribution")
plt.show()
```
